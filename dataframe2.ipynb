# boilplate code

```python
df = pd.read_csv('data/employees.csv', parse_dates=['Start Date', 'Last Login Time'])
df['Gender'] = df['Gender'].astype('category')
df['Senior Management'] = df['Senior Management'].astype('bool')
df.info()
```

# filter dataframe

```python
df.head()
```

```python
mask = df['Gender'] == 'Male'
df[mask]
```

```python
mask = df['Senior Management'] # boolean Series can be used directly
df[mask]
```

```python
mask = df['Team'] != 'Marketing'
df[mask]
```

```python
mask = df['Salary'] > 110_000
df[mask]
```

```python
mask = df['Bonus %'] < 1.5
df[mask].head()
```

```python
mask = df['Start Date'] < '1995-01-01'
df[mask]
```
