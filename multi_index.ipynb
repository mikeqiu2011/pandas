# swaplevel

```python
df = df.transpose()
df.head()
```

```python
df.swaplevel()
df.swaplevel('Date', 'Country')
df.swaplevel('Country', 'Date')
df.swaplevel(0,1)
df.swaplevel(1,0)
```

# pivot

```python
sales = pd.read_csv('data/salesmen.csv', parse_dates=['Date'])
sales.head(3)
```

```python
sales.info()
```

```python
sales.nunique()
```

```python
sales['Salesman'] = sales['Salesman'].astype('category')
sales.info()
```

> useful to condense dataframe

```python
# only ideal when you only have small number of values of the "column"
sales.pivot(index='Date', columns='Salesman', values='Revenue')
```

```python
# not desirable as too many value in "date"!!
sales.pivot(index='Salesman', columns='Date', values='Revenue')
```

# pivot_table()

> similar to pivot() but more focus on aggregate data

```python
df = pd.read_csv('data/foods.csv')
df.head(3)
```

```python
df.info()
```

```python
df.nunique()
```

```python
df.pivot_table(values='Spend', index='Gender', aggfunc=['mean', 'sum', 'count'])
```

```python
df.pivot_table(values='Spend', index=['Gender', 'Item'], aggfunc=['mean'])
```

```python
df.pivot_table(values='Spend', index=['Gender', 'Item'], columns='City', aggfunc=['mean'])
```

```python
df.pivot_table(values='Spend', index=['Gender', 'Item'], columns=['Frequency', 'City'], aggfunc=['mean'])
```
